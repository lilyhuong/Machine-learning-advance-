```python
# Importing required modules
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder 
```

```python
# Let us create the Data-Frame for above 
df = pd.DataFrame({'LikesExercising':[False,False,False,True,False,True,True,True,True],
                'GotoGym':[True,True,True,True,True,False,True,False,False],
                 'DrivesCar':[True,False,False,True,True,False,True,False,True],
                 'Age':[14,15,16,26,36,50,69,72,74]})
```

```python
df
```

```python
LE = LabelEncoder()
df['LikesExercising'] = LE.fit_transform(df['LikesExercising'])
df['GotoGym'] = LE.fit_transform(df['GotoGym'])
df['DrivesCar'] = LE.fit_transform(df['DrivesCar'])
```
